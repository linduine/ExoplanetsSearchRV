```python
#populating space with libraries: numpy (calculations), matplotlib (drawing plots), pyswarm -- Particle Swarm Optimization
%pylab inline
from pyswarm import pso
```

```python
data = loadtxt('planetXS.dat') # loading files, for examples planetXS.dat, with only one planet injected

time = np.zeros(len(data), dtype=float64)
value = np.zeros(len(data), dtype=float64)
errorBr = np.zeros(len(data), dtype=float64)

for i in range(0, len(data)): #populating numpy arrays with values from the file
    time[i] = float(data[i][0])
    value[i] = float(data[i][1])
    errorBr[i] = float(data[i][2])
    
plot(time, value) #plotting data 
timespan = time[-1]-time[0] #calculating timespan to be used in PSO
```

```python
def eccNewton(t, tp, P, e): # calculating eccentricity using Newton method
    sEccNewton = np.zeros(len(t))
    Mtime = (2.*pi*(t - tp))/P 
    eccNewton = Mtime
    while(norm(eccNewton-sEccNewton)>0.001):
        sEccNewton = eccNewton
        eccNewton = eccNewton + (Mtime + e*sin(eccNewton) - eccNewton)/(1-e*cos(eccNewton))
    nuT = 2.*arctan(sqrt((1.+e)/(1.-e))*tan(eccNewton/2.))
    
    return nuT
```

```python
def f(t, gamma, K, e, omega, P, tp): # calculating radial velocity function
    nuT = eccNewton(t, tp, P, e)
    VrT = gamma + K*(cos(nuT+omega) + e*cos(omega))
    
    return VrT
```

```python
def forPSO(P0, *args): # calculating -log of likelihood function to pass into PSO function 
    t, V, errBr = args
    gamma, K, e, omega, P, tp = P0
    P = exp(P) # since timespan can be big, we use log scale for period in input parameters, thus to get real period we need
    # to take exponent of the input parameter P
    nuT = eccNewton(t, tp, P, e)
    VrT = gamma + K*(cos(nuT+omega) + e*cos(omega))
    residue = (V-VrT)/errBr
    expV = -0.5*(residue**2)
    logL = -1*(sum(log(1./(sqrt(2.*pi)*errBr))) + sum(expV))
    return logL
```

```python
lb = [-1., 0., 0., 0., log(0.1), 0.] #lower bound for PSO
ub = [1., .1, 0.9, 2*pi, log(timespan), timespan] # upper bound for PSO
args = (time, value, errorBr)
```

```python
xopt, fopt = pso(forPSO, lb, ub, args=args, swarmsize=150, maxiter=10000, minstep = 1e-15, minfunc=1e-15) #minimizing return
#value of forPSO function, which is -log of our likelihood function
xopt[4] = exp(xopt[4]) # taking exponent of the P value to get true period
```

```python
print(xopt)
```

```python
# Drawing plot
phase = (time - xopt[5])%xopt[4]/xopt[4]
xx = linspace(xopt[5], xopt[5] + xopt[4], 1000)
yy = f(xx, xopt[0], xopt[1], xopt[2], xopt[3], xopt[4], xopt[5])
phase2 = (xx - xopt[5])%xopt[4]/xopt[4]
yy = yy[argsort(phase2)]
phase2 = sort(phase2)
errorbar(phase, value, xerr=0., yerr=errorBr, fmt = 'or', label="Data") #plotting data points
ylabel('Radial velocity, [km/s]')
xlabel('Phase')
plot(phase2, yy, '-k', label="Model Fit") #plotting fit curve
legend()
print("gamma: " + str(xopt[0]) + " K: " + str(xopt[1]) + " e: " + str(xopt[2]) + " omega:" + str(xopt[3]) + 
      " Period : " + str(xopt[4]) + " tp : " + str(xopt[5])) #printing corresponding values of the minimized function
```
