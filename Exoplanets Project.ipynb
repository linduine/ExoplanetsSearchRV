```python
#Parameters
P = 8.
t=np.arange(0, P, 0.001)
e = 0.5
K = 0.2
omega = pi/2.
tp = 1.
gamma = 0.
p0 = [gamma, K, e, omega, P, tp]
```

```python
#eccentricty for all time --> M(t) is a vector
def eccNewton(t, tp, P, e):
    sEccNewton = np.zeros(len(t))
    
    Mtime = (2.*pi*(t - tp))/P #N(time)
    eccNewton = Mtime
    
    while(norm(eccNewton-sEccNewton)>0.01):
        sEccNewton = eccNewton
        eccNewton = eccNewton + (Mtime - eccNewton + e*sin(eccNewton))/(1-e*cos(eccNewton))
    nuT = 2.*arctan(sqrt((1+e)/(1-e))*tan(eccNewton/2.))
    
    return nuT

def f(t, gamma, K, e, omega, P, tp):
    nuT = eccNewton(t, tp, P, e)
    VrT = gamma + K*(cos(nuT+omega) + e*cos(omega))
    
    return VrT

def f1(P0, t, V, sigma):
   # gamma, K, e, omega, P, tp = P0
    #nuT = eccNewton(t, P0[6], P0[5], P0[3])
    # add weight in here
    VrT = f(t, P0[0], P0[1], P0[2], P0[3], P0[4], P0[5])
    return (V-VrT)/sigma
```

```python
%pylab inline
from pyswarm import pso
from time import * #ts = time.
```

```python
#eccentricty for all time --> M(t) is a vector
data = loadtxt('BT1_wn.dat') # other files : 

time = np.zeros(len(data), dtype=float64)
value = np.zeros(len(data), dtype=float64)
errorBar = np.zeros(len(data), dtype=float64)
for i in range(0, len(data)):
    time[i] = float(data[i][0])
    value[i] = float(data[i][1])
    errorBar[i] = float(data[i][2])
    
def eccNewton(t, tp, P, e):
    sEccNewton = np.zeros(len(t))
    
    Mtime = (2.*pi*(t - tp))/P #N(time)
    eccNewton = Mtime
    
    while(norm(eccNewton-sEccNewton)>0.01):
        sEccNewton = eccNewton
        eccNewton = eccNewton + (Mtime - eccNewton + e*sin(eccNewton))/(1-e*cos(eccNewton))
    nuT = 2.*arctan(sqrt((1+e)/(1-e))*tan(eccNewton/2.))
    
    return nuT

def f(t, gamma, K, e, omega, P, tp):
    nuT = eccNewton(t, tp, P, e)
    VrT = gamma + K*(cos(nuT+omega) + e*cos(omega))
    
    return VrT

def f1(P0, t, V, sigma):
   # gamma, K, e, omega, P, tp = P0
    #nuT = eccNewton(t, P0[6], P0[5], P0[3])
    # add weight in here
    VrT = f(t, P0[0], P0[1], P0[2], P0[3], P0[4], P0[5])
    return (V-VrT)/sigma

def my_function(P0, t, V, errorBar):
    expV = -0.5*f1(P0, t, V, errorBar)**2
    #L = 
    logL = sum(log(1/(sqrt(2*pi)*errorBar))) + sum(expV)
    return -logL #because pso searches for minimum when we need maximum likelihood

#should be run on posterior, not the likelihood, but priors are uniformly distributed, so we can skip using them cause they have
# a constant value
lb = [-1., 0., 0., 0., 0., 0.]
ub = [1, 100, 0.9, 2*pi, 200., 200.] #doesn't get stuck with ecc around 1.

#valueTheory = f(time, 0., 0., 0., 0., 100., 5.)
#print(my_function(valueTheory, value, errorBar))
#stats.uniform.pdf
xopt, fopt = pso(my_function, lb, ub, swarmsize=100, maxiter=80000, minstep=1e-12, minfunc=1e-12, args=[time, value, errorBar]) # ,
#print(xopt)
#best result so far for BT4: 
#gamma: -0.00558327297753 K: 0.0 e: 0.525419011369 omega:5.48604892388 Period : 118.110765831 tp : 24.7685573208
#swarmsize=300, maxiter=40000, minstep=1e-14, minfunc=1e-15,
```

```python
#Phase = (t-tp)/P[]
#phase = (t - tp)$P
phase = (time - xopt[5])%xopt[4]/xopt[4] #other way around ((t-tp)/P)%1
#value = value[argsort(phase)]
#errorBar = errorBar[argsort(phase)]
#the same for sigma
value = value[argsort(phase)]
errorBar = errorBar[argsort(phase)]
phase = sort(phase)
#plot(phase, value)
xx = linspace(xopt[5], xopt[5] + xopt[4], 1000)
yy= f(xx, xopt[0], xopt[1], xopt[2], xopt[3], xopt[4], xopt[5])
phase2 = (xx - xopt[5])%xopt[4]/xopt[4]
yy = yy[argsort(phase2)]
phase2 = sort(phase2)
errorbar(phase, value, xerr=0., yerr=errorBar, fmt = 'or')
plot(phase2, yy, '-k')
print("gamma: " + str(xopt[0]) + " K: " + str(xopt[1]) + " e: " + str(xopt[2]) + " omega:" + str(xopt[3]) + 
      " Period : " + str(xopt[4]) + " tp : " + str(xopt[5]))
```
